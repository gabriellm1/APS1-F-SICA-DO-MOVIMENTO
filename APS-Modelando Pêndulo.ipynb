```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import arange
from math import *
```

```python
# Modelando o pêndulo com ângulo de -14,7 graus

#parâmetros
g=9.8 
dt=0.001
l=0.64 #m
m=0.023 #kg
tempo=arange(0,2,dt)

# condições iniciais
𝜃_inicial = 194.7 #ângulo alterado para se comparar com o programa Tracker que inverteu os lados
𝜔_inicial = 0
IC=[radians(𝜃_inicial),𝜔_inicial]




def EqDif(Solucao, t):
    𝜃=Solucao[0]
    𝜔=Solucao[1]
    d𝜃dt= 𝜔
    d𝜔dt=  -g*cos(𝜃)/l 
    return [d𝜃dt, d𝜔dt]



Solucao=odeint(EqDif, IC, tempo)

x1=[]
y1=[]
teta1=[]
T1=[]

for i in range(0,len(Solucao[:,0])):
    𝜃=(Solucao[:,0])[i]
    𝜔=(Solucao[:,1])[i]
    x1.append(l*cos(𝜃))
    y1.append(l*sin(𝜃))
    teta1.append(𝜃)
    T1.append(m*g*sin(𝜃)-m*𝜔**2*l)
    


plt.plot(tempo, teta1,'g')
plt.xlabel('Tempo (s)')
plt.ylabel('Ângulo teta(rad)')
plt.grid(True)
plt.axis('equal')
plt.show()

plt.plot(tempo, x1,'g')
plt.xlabel('Tempo (s)')
plt.ylabel('X')
plt.grid(True)
plt.axis('equal')
plt.show()

plt.plot(tempo, y1,'black')
plt.xlabel('Tempo (s)')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show()

plt.plot(x1, y1,'r')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show()



plt.plot(tempo, T1, 'b', label='T(t)')
plt.xlabel('Tempo (s)')
plt.ylabel('Tração(N)')
plt.grid(True)
plt.show()
```

```python
# Modelando o pêndulo com ângulo de -57,7 graus

# Condições iniciais
𝜃_inicial = 237.7 #ângulo alterado para se comparar com o programa Tracker que inverteu os lados
𝜔_inicial = 0
IC=[radians(𝜃_inicial),𝜔_inicial]



Solucao2=odeint(EqDif, IC, tempo)

x2=[]
y2=[]
teta2=[]
T2=[]

for i in range(0,len(Solucao2[:,0])):
    𝜃=(Solucao2[:,0])[i]
    𝜔=(Solucao2[:,1])[i]
    x2.append(l*cos(𝜃))
    y2.append(l*sin(𝜃))
    teta2.append(𝜃)
    T2.append(m*g*sin(𝜃)-m*𝜔**2*l)
    

plt.plot(tempo, teta2,'g')
plt.xlabel('Tempo (s)')
plt.ylabel('Ângulo teta(rad)')
plt.grid(True)
plt.axis('equal')
plt.show()


plt.plot(tempo, x2,'g')
plt.xlabel('Tempo (s)')
plt.ylabel('X')
plt.grid(True)
plt.axis('equal')
plt.show()

plt.plot(tempo, y2,'black')
plt.xlabel('Tempo (s)')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show()

plt.plot(x2, y2,'r')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show()


plt.plot(tempo, T2, 'b', label='T(t)')
plt.xlabel('Tempo (s)')
plt.ylabel('Tração(N)')
plt.grid(True)
plt.show()
```
